# Error Bars with Plotly Express

## Express

```python
import plotly.express as px

df = px.data.iris()
df['e'] = df['sepal_width'] / 100
fig = px.scatter(df,
                 x='sepal_width', y='sepal_length',
                 color='species',
                 error_x='e', error_y='e',
                 width=800, height=400)
fig.show()
```

## graph_objects

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=go.Scatter(x=[0, 1, 2],
                    y=[6, 10, 2],
                    mode='markers', 
                    error_y=dict(
                        type='data', # value of error bar given in data coordinates
                        symmetric=False,
                        array=[0.1, 0.2, 0.1],
                        arrayminus=[0.2, 0.4, 1],
                        visible=True))
)
fig.update_layout(width=800, height=400)
fig.show()
```
