```python
from spektral.datasets import citation
# A is adjacency matrix (N x N)
# X is node characteristics matrix (N x F)
# y is label matrix (N x #n_classes)
A, X, y, train_mask, val_mask, test_mask = citation.load_data('cora')

N = A.shape[0]
F = X.shape[-1]
n_classes = y.shape[-1]
```

```python
from spektral.layers import GraphConv
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dropout
```

```python
## construct model
X_in = Input(shape=[F, ])
A_in = Input(shape=[N, ], sparse=True)

X_1 = GraphConv(16, 'relu')([X_in, A_in])
X_1 = Dropout(0.5)(X_1)
X_2 = GraphConv(n_classes, 'softmax')([X_1, A_in])

model = Model(inputs=[X_in, A_in], outputs=X_2)
```

```python
## compile GNN
A = GraphConv.preprocess(A).astype('f4')

model.compile(optimizer='adam', \
              loss='categorical_crossentropy', \
              weighted_metrics=['acc'])
model.summary()
```

```python
## prepare data
X = X.toarray()
A = A.astype('f4')
validation_data = ([X, A], y, val_mask)
```

```python
## train GNN
model.fit([X, A], y, \
          epochs=200, \
          sample_weight=train_mask, \
          validation_data=validation_data, \
          batch_size=N, \
          shuffle=False)
```

```python
## evaluate model
eval_results = model.evaluate([X, A], y, \
                              sample_weight=test_mask, \
                              batch_size=N)
print('Done. \n''Test loss: {}\n''Test accuracy: {}'.format(*eval_results))
```
