## CNN Architectures

### LeNet-5

| Layer |   Type   | Out_channel | out_size | Kernel_size | Stride | Activation |
|:-----:|:--------:|:-----------:|:--------:|:-----------:|:------:|:----------:|
|  Out  |    FC    |      -      |    10    |      -      |    -   |     RBF    |
|   F6  |    FC    |      -      |    84    |      -      |    -   |    tanh    |
|   C5  |   Conv   |     120     |   1 x 1  |    5 x 5    |    1   |    tanh    |
|   S4  | Avg Pool |      16     |   5 x 5  |    2 x 2    |    2   |    tanh    |
|   C3  |   Conv   |      16     |  10 x 10 |    5 x 5    |    1   |    tanh    |
|   S2  | Avg Pool |      6      |  14 x 14 |    2 x 2    |    2   |    tanh    |
|   C1  |   Conv   |      6      |  28 x 28 |    5 x 5    |    1   |    tanh    |
|   In  |   Input  |      1      |  32 x 32 |      -      |    -   |      -     |

- Input size is 28 x 28, but applying zero-padding to 32 x 32. Rest of layers padding='VALID'
- AvgPooling first calculate mean of current window, then multiple mean by $\beta_1$, then add a bias term $\beta_0$. (Like the second step in BatchNorm - linearScaling)
- Most filter in C3 Conv is 3/4 group convolution (in Pytorch Conv `group` parameter)
- Output layer is calculating the Euclidian distanace between inputs vector and output_weights (1x84 and 84x10) to measure hot much the image belongs to a particular digit class.

## Implementing a ResNet-34 CNN Using Keras

```python
from tensorflow import keras
```

```python
class ResidualUnit(keras.layers.Layer):
    def __init__(self, filters, strides=1, activation='relu', **kwargs):
        super(ResidualUnit, self).__init__(**kwargs)
        self.activation = keras.activations.get(activation)
        self.main_layers = [
            keras.layers.Conv2D(filters, kernel_size=(3,3), strides=strides,
                                padding='SAME', use_bias=False),
            keras.layers.BatchNormalization(),
            self.activation,
            keras.layers.Conv2D(filters, kernel_size=(3,3), strides=1,
                                padding="SAME", use_bias=False),
            keras.layers.BatchNormalization()
        ]
        self.skip_layers = []
        if strides > 1:
            self.skip_layers = [
                keras.layers.Conv2D(filters, kernel_size=(1,1), strides=strides,
                                    padding='SAME', use_bias=False),
                keras.layers.BatchNormalization()
            ]
        
    def call(self, inputs):
        Z = inputs
        for layer in self.main_layers:
            Z = layer(Z)
        skip_Z = inputs
        for layer in self.skip_layers:
            skip_Z = layer(skip_Z)
        return self.activation(Z + skip_Z)
```

```python
model = keras.Sequential()
# 1st conv layer & max pooling
model.add(keras.layers.Conv2D(64, kernel_size=(7,7), strides=2, input_shape=(224, 224, 3), 
                              padding="SAME", use_bias=False))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Activation("relu"))
model.add(keras.layers.MaxPool2D(pool_size=(3,3), strides=2, padding="SAME"))

# add residual blocks
prev_filters = 64
for filters in [64]*3 + [128]*4 + [256]*6 + [512]*3:
    strides = 1 if filters == prev_filters else 2
    model.add(ResidualUnit(filters, strides=strides))
    prev_filters = filters

# add end layers
model.add(keras.layers.GlobalAveragePooling2D())
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(10, activation='softmax'))
```

## Pretrained Models for Transfer Learning

Use Xception model

```python
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow import keras
```

```python
dataset, info = tfds.load('tf_flowers', as_supervised=True, with_info=True)
```

```python
dataset_size = info.splits['train'].num_examples
class_name = info.features['label'].names
n_classes = info.features['label'].num_classes
```

```python
test_set, valid_set, train_set = tfds.load('tf_flowers', split=['train[:10%]', 'train[10%:25%]', 'train[25%:]'],
                                          as_supervised=True, with_info=False)
```

```python
def preprocess(image, label):
    resized_image = tf.image.resize(image, (224, 244))
    final_image = keras.applications.xception.preprocess_input(resized_image)
    return final_image, label
```

```python
train_set.take(1)
```

```python
batch_size = 4
train_set = train_set.shuffle(1000).map(preprocess).batch(batch_size).prefetch(1)
valid_set = valid_set.map(preprocess).batch(batch_size).prefetch(1)
test_set = test_set.map(preprocess).batch(batch_size).prefetch(1)
```

```python
train_set.take(1) # now the batch dimension has added
```

```python
base_model = keras.applications.xception.Xception(weights='imagenet', include_top=False)
```

```python
avg = keras.layers.GlobalAvgPool2D()(base_model.output)
output = keras.layers.Dense(n_classes, activation='softmax')(avg)
model = keras.models.Model(inputs=[base_model.input], outputs=[output])
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

decay:

$$
lr = lr * (1 / (1 + self.decay * self.iterations))
$$

```python
optimizer = keras.optimizers.SGD(0.2, momentum=0.9, decay=0.01)
```

```python
model.compile(loss=keras.losses.sparse_categorical_crossentropy, optimizer=optimizer, metrics=[tf.keras.metrics.SparseCategoricalAccuracy()])
```

```python
history = model.fit(train_set, epochs=5, validation_data=valid_set)
```

```python
import pandas as pd
pd.DataFrame(history.history).plot()
```

```python
# you can see the val_acc is getting stuck on 75-80%, which means the FC layers is trained well, now we
# unfreeze the pretrained model to fine-tune it
for layer in base_model.layers:
    layer.trainable = True
```

```python
optimizer = keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, decay=0.001)
model.compile(loss=keras.losses.sparse_categorical_crossentropy, optimizer=optimizer, metrics=[keras.metrics.SparseCategoricalAccuracy()])
```

```python
history = model.fit(train_set, epochs=5, validation_data=valid_set)
```

```python
import pandas as pd
pd.DataFrame(history.history).plot()
```

## Classification and Localization

```python
# output the bounding box along with probabilities
avg = keras.layers.GlobalAvgPool2D()(base_model.output)
class_output = keras.layers.Dense(n_classes, activation='softmax')(avg)
loc_output = keras.layers.Dense(4)(avg)
model = keras.models.Model(inputs=[base_model.input], outputs=[class_output, loc_output])
model.compile(loss=[keras.losses.sparse_categorical_crossentropy, keras.losses.mse], loss_weights=[0.8, 0.2], optimizer=optimizer, metrics=[keras.metrics.sparse_categorical_accuracy])
```
